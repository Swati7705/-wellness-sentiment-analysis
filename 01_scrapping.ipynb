```python
from google_play_scraper import reviews, Sort
import pandas as pd
import time

print("Libraries imported successfully!")
```

```python
# Cell 2: Define which apps we want to scrape
apps_to_scrape = {
    'Habitica'      : 'com.habitrpg.android.habitica',
    'Headspace'     : 'com.getsomeheadspace.android',
    'MyFitnessPal'  : 'com.myfitnesspal.android',
    'Fabulous'      : 'co.thefabulous.app',
    'Finch'         : 'com.finchcare.finch'
}

print("Apps defined:", list(apps_to_scrape.keys()))
```

```python
# Cell 3: Scrape reviews from all apps
all_reviews = []

for app_name, app_id in apps_to_scrape.items():
    print(f"Scraping {app_name}...")
    
    try:
        result, _ = reviews(
            app_id,
            lang='en',
            country='us',
            sort=Sort.NEWEST,
            count=300   # 300 reviews per app
        )
        
        for review in result:
            review['app_name'] = app_name
        
        all_reviews.extend(result)
        print(f"  ✓ Got {len(result)} reviews")
        time.sleep(2)  # pause to be polite to the server
        
    except Exception as e:
        print(f"  ✗ Error: {e}")

print(f"\nTotal reviews collected: {len(all_reviews)}")
```

```python
# Cell 4: Convert to DataFrame and save
df_raw = pd.DataFrame(all_reviews)

print("Shape of data:", df_raw.shape)
print("\nColumns:", df_raw.columns.tolist())
print("\nFirst 3 rows:")
print(df_raw[['app_name', 'content', 'score']].head(3))

# Create a data folder right here where the notebook is
import os
os.makedirs('data', exist_ok=True)

# Save the raw data
df_raw.to_csv('data/raw_reviews.csv', index=False)
print("\n✓ Data saved successfully!")
print("Saved to:", os.path.abspath('data/raw_reviews.csv'))
```
